```python
#sum of numbers form 1 to n.

def sumOfNum(n):
    result=0
    for i in range(1,n+1,1):
        result+=i
    return result
n= int(input("value : "))
result = sumOfNum(n)
print("total sum is : ",result)
```

```python
#calculate n factorial

def facto(n):
    result=1
    for i in range(1,n+1):
        result*=i
    return result 

n = int(input(" value : "))
result = facto(n)
print(f"factorial of given {n} is : ",result)
```

```python
def alp(num):
    digg=0
    while(num>0):
        last_dig =num%10
        num =int(num / 10)
        digg += int(last_dig)
    return digg


num = int(input("number :"))
result = alp(num)
print ("main result is :" ,result) 
```

```python
#homework
def primeOrNOt(num):
    for i in range(2,num+1):
        if(num%i==0):
            return "Not Prime"
        else:
            return "Prime"
num= int(input("enter num: "))
result= primeOrNOt(num)
print(f"{num} is {result}")
```
