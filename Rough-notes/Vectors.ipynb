## Vectors

### Adding element

```python
v = [1, 2, 3]

v.append(4)           # add at end        → [1,2,3,4]
v.insert(1, 99)       # add at index 1    → [1,99,2,3,4]
v.extend([5,6])       # add multiple      → [1,99,2,3,4,5,6]
```

### Remove

```python
v.pop()               # remove last         → returns removed element
v.pop(1)              # remove at index 1   → returns removed element
v.remove(99)          # remove by value     → removes first occurrence
v.clear()             # remove everything   → []
```

### Searching

```python
v = [10, 20, 30, 20]

v.index(20)           # first index of 20  → 1
v.count(20)           # how many 20s       → 2
20 in v               # exists?            → True
```

### Ordering

```python
v = [3, 1, 4, 2]

v.sort()              # ascending          → [1,2,3,4]
v.sort(reverse=True)  # descending         → [4,3,2,1]
v.reverse()           # flip the list      → [2,3,4,1] (just reverses, not sort)

sorted(v)             # returns NEW sorted list, original unchanged
```

### Slicing

```python
v = [10, 20, 30, 40, 50]

v[1:3]                # index 1 to 2      → [20,30]
v[:3]                 # start to index 2  → [10,20,30]
v[2:]                 # index 2 to end    → [30,40,50]
v[::2]                # every 2nd element → [10,30,50]
v[::-1]               # reverse           → [50,40,30,20,10]
```
