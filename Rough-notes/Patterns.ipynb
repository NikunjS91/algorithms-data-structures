```python
#in this case we wanted the sequence form 1 to 9
#we never want to reset the value of num so we put it out of loops
#x-axis we need is n and y is also n so loop works on range of n.
n=int(input("enter value: "))
num=0
for i in range(n):
    print() #after printing values its like we are hiting Enter button.
    for j in range(n):
        num+=1 #we are incrementing it with combination of both loop with is 3*3 = 9
        print(num," ",end="")
print()
print("End")
```
