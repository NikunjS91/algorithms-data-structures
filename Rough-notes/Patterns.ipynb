```python
## pyramid

n =int(input("value : "))
for i in range(n):
    print()
    for j in range(n-i-1):
        print(" ",end="")
    for j in range(0,i+1,1):
        print(j+1,end="")
    for j in range(i,0,-1):
        print(j,end="")
```

```python
n =int(input("value : "))
for i in range(n):
    print()
    for j in range(n-i-1):
        print(" ",end="")
    print("*",end="")
    if(i!=0):
        for j in range(2*i-1):
            print(" ",end="")
        print("*",end="")
for i in range(n-1):
    print()
    for j in range(i+1):
        print(" ",end="")
    print("*",end="")
    if(i!=(n-2)):
        for j in range(0,2*(n-i)-5,1):
            print(" ",end="")
        print("*",end="")

    
```

```python
##butterfly

n = int(input("value : "))
for i in range (n):
    print()
    for j in range(i+1):
        print("*", end="")
    if(i!=n):
        for j in range((n*2)-(i*2+2)):
            print(" ",end="")
    for j in range(i+1):
            print("*",end="")
for i in range(n):
    print()
    for j in range(n-i):
        print("*",end="")
    if(i!=0):
        for j in range(i*2):
             print(" ",end="")
    for j in range(n-i):
         print("*",end="")
```
