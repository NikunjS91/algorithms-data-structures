```python
def powerOf2(num):
    if num <= 0:
        return "not a power of 2."
    while num > 1:
        remainder = num % 2
        if remainder != 0:      # found a remainder, stop
            return "not a power of 2."
        num = num // 2          # keep dividing
    return "it is a power of 2."

num = int(input("enter number : "))
result = powerOf2(num)
print(result)
```
