```python

def pairSum(arr,target):
    size = len(arr)
    i=0
    j=size-1
    while(i<j):
        sum=arr[i]+arr[j]
        if(sum<target):
            i+=1
        elif(target<sum):
            j-=1
        else:
            return (f"sum of there element {arr[i],arr[j]} is : {target}")


arr=[2,7,11,15]
target=13
result=pairSum(arr,target)
print(result)
        
```
