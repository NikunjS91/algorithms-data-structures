```python
arr=[-1,1,2,3,4,-5]
n=len(arr)
max_sum=0
for st in range(n):
    arr_sum=0
    for end in range(st,n,+1):
        arr_sum+=arr[end]
        max_sum = max(arr_sum,max_sum)
print(max_sum)
```
