```python

def pairSum(arr,target):
    size = len(arr)
    i=0
    j=size-1
    while(i<j):
        sum=arr[i]+arr[j]
        if(sum<target):
            i+=1
        elif(target<sum):
            j-=1
        else:
            return (f"sum of there element {arr[i],arr[j]} is : {target}")


arr=[2,7,11,15]
target=13
result=pairSum(arr,target)
print(result)
        
```

```python
def majority(arr):
    size=len(arr)
    max_count=0
    for i in range(size):
        count=0
        for j in range(i,size):
            if(arr[i]==arr[j]):
                count+=1
                if(max_count<count):
                    max_count=count
    print(i,j,":",max_count)

    

arr=[1,2,2,1,1]
result=majority(arr)
print(result)
```
