#decimal to binary


```python

def decToBi(num):
    rem=0
    power=1
    ans=0
    while(num>0):
        rem=num%2
        num//=2
        ans+=(rem*power)
        power*=10
        #print(power)
    return ans

num=int(input("enter Decimal number : "))
res=decToBi(num)
print(res)
```

#Binary to Decimal


```python
def biToDec(num):
    ans=0
    power=1
    rem=0
    while(num>0):
        rem=num%10
        num//=10
        ans+=(rem*power)
        power*=2
    return ans

num=int(input("enter Decimal number : "))
res=biToDec(num)
print(res)
```
