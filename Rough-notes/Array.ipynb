#pass by referance

```python
#linear search
def indexOf(mylist,target):
    for i in range(len(mylist)):
        if(mylist[i]==target):
           return i
    else:
        return -1


mylist = [50, 15, 21, 1, 22, 24]
target = 15
result = indexOf(mylist,target)
print("index is : ",result)
```

```python
mylist = [50, 15, 21, 1, 22, 24,12]
start=0
end=len(mylist)-1
while(start<end):
    mylist[start],mylist[end]=mylist[end],mylist[start]
    start+=1
    end-=1
print(mylist)
```

```python
def prodAndSum(arr):
    size=len(arr)
    sum=0
    prod=1
    for i in range(size):
        sum+=arr[i]
        prod*=arr[i]
    return sum,prod

mylist = [50, 15, 21, 1, 22, 24,12]
result=prodAndSum(mylist)
print(f"Product and Sum of given array us {result}")
```

```python
def max(arr):
    size=len(arr)
    start=0
    end=size-1
    while(start<=end):
        if(arr[start]<arr[end]):
            arr[start],arr[end]=arr[end],arr[start]
            start+=1
            end-=1
        else:
            break
    return arr

def mini(arr1):
    size=len(arr1)
    start=0
    end=size-1
    while(start<=end):
        if(arr1[start]>arr1[end]):
            arr1[start],arr1[end]=arr1[end],arr1[start]
            start+=1
            end-=1
        else:
            break
    return arr1

mylist = [1,2,3,4,5]
result=max(mylist)
print(f"{result}")
minresult=mini(result)
print(f"{minresult}")
```
